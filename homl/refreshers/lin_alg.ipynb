## Plotting vectors

```python
import matplotlib.pyplot as plt
```

```python
u = np.array([3, 5])
v = np.array([5, 4])
```

```python
x_coords, y_coords = zip(u, v)
plt.scatter(x_coords, y_coords, color=['r', 'b'])
plt.axis([0, 6, 0, 10])
plt.grid()
plt.show()
```
